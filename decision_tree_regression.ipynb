```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv('golf.txt')
df.head(15)
```

```
def statiscs_cal(hour_ser):
    ser_mean = hour_ser.mean()
    ser_std = np.sqrt(np.sum(np.power((hour_ser - ser_mean), 2)) / len(hour_ser))
    return ser_mean, ser_std
```

```
grp = df.groupby('Outlook')

for outlook, group in grp:
    print(outlook)
    print(group)
    print()
```

```
def group_sd_cal(df, attribute):
    grp = df.groupby(attribute)
    sc = lambda x: np.std(x, ddof=0) 
    #grp = df.groupby(attribute).agg(['count', x])
    print(grp.transform(sc))
    #print(grp['Decision'])
    print()
```

```
attributes = ['Outlook', 'Temp', 'Humidity', 'Wind']
for attribute in attributes:
    group_sd_cal(df, attribute)
```
